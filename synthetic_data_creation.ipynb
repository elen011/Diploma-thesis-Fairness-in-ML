```python
import numpy as np
import pandas as pd
import glob
```

```python
def generate_training_set(m=1000):
    # Generate treatment assignment T (0 or 1 with equal probability)
    T = np.array([0] * (m//2) + [1] * (m//2))
    
    # Generate Y based on T
    Y = np.where(T == 0, (np.random.normal(1, 1, size=m)),(np.random.normal(-1, 1, size=m)))
    
    # Generate X based on Y
    X = Y + np.random.normal(0, 1, m)
    
    # Create DataFrame
    df = pd.DataFrame({'X': X, 'Y': Y, 'T': T})
    
    return df
```

```python
def generate_multiple_datasets(n=10000, m=1000):
    for i in range(n):
        df = generate_training_set(m)
        df.to_csv(f"synthetic_data2_{i+1}.csv", index=False)
        print(f"Created: synthetic_data2_{i+1}.csv")  # This will confirm the files being created

generate_multiple_datasets(10000, 1000)
```

```python
import os
# Get a list of all CSV files in the directory
file_list = glob.glob("synthetic_data2_*.csv") 

# Initialize an empty list to store dataframes
dataframes = []

# Loop through each file, read it, and append to the list
for i, file in enumerate(file_list):
    df = pd.read_csv(file)
    df["Dataset_ID"] = i + 1  # Add a column to indicate which dataset it came from
    dataframes.append(df)

    # Delete the file after it is processed
    os.remove(file)
    print(f"Deleted {file}")

# Concatenate all dataframes into one large dataframe
merged_df = pd.concat(dataframes, ignore_index=True)

# Save the merged dataframe as a single CSV file
merged_df.to_csv("all_synthetic_data2.csv", index=False)

print("Merged CSV file saved as 'all_synthetic_data2.csv'")

```
